# Exercise 3: Drawing shapes and text on an image

https://docs.opencv.org/4.6.0/dc/da5/tutorial_py_drawing_functions.html

In this exercise you will learn how to draw different geometric shapes with the following OpenCV functions:
- `cv.line()` draws a line on an image.
- `cv.rectangle()` draws a rectangle on an image.
- `cv.circle()` draws a circle on an image.
- `cv.ellipse()` draws an ellipse on an image.
- `cv.polylines()` draws one, or more, polygons on an image.
- `cv.putText()` draws text on an image.

The functions above have som common arguments as given below:
- *img* : The image where you want to draw the shapes.
- *color* : The shape's color. For BGR, pass in a tuple, e.g. (255,0,0) for blue. For grayscale, pass in a scalar value.
- *thickness* : The line thickness (default = 1). If -1 is passed, closed figures will be filled in with the color above.
- *lineType* : The line type, e.g. 8-connected `cv.LINE_8` (default) or anti-aliased `cv.LINE_AA`.

You will need the following python packages: `pip install numpy matplotlib ipykernel opencv-python`

**Note 1**: to get the *docstring* (i.e. a description) for a function, just add a question mark '?' after the function name, and run the resulting expression in a code cell, e.g. `cv.line?`.

**Note 2**: let's use matplotlib for plotting the images (instead of cv.imshow(), cv.waitKey() and cv.destroyAllWindows()), but remember to reverse the order of the color channels from (b,g,r) to (r,g,b) e.g. `img[:,:,::-1]` before doing so (and converting back again before using any OpenCV functions on an image).

As a first step, let's import the python modules we need.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## Drawing lines

The `cv.line()` function expects the input arguments below:
- The image to draw on.
- The start coordinates (column, row).
- The end coordinates (column, row).
- A color: (b,g,r) values for a BGR image, a scalar value for a gray scale image.
- The line thinckness in pixels (px).

```python
# Create a black 512x512 image, with 3 color channels, each with a depth of 8-bits
img = np.zeros((512,512,3), np.uint8)

# Draw a diagonal blue (b,g,r=255,0,0) line from pixel (0,0) to (511,511) with a line thickness of 5 px (pixels)
cv.line(img, (0,0), (511,511), (255,0,0), 5)

# Display the image with matplotlib (remembering to reverse the color channels)
plt.imshow(img[:,:,::-1]);
plt.xticks([]); plt.yticks([]);
plt.show()

# Notice that the pixels have data type "unit8" (8-bit unsigned int)
# with values 0-255 since we chose this when creating the image
print(f'\nTop 10x10 pixels of the blue channel:\n\n{img[0:10,0:10,0]}')
```

## Drawing rectangles

The `cv.rectangle()` function expects the input arguments below:
- The image to draw on.
- The rectangle's top left corner (column, row).
- The rectangle's bottom left corner (column, row).
- A color: (b,g,r) values for a BGR image, a scalar value for a gray scale image.
- The line thinckness in pixels (px), or -1 to fill the rectangle with the chosen color.

```python
# Draw a green (b,g,r=0,255,0) rectangle with topleft (column,row)=(384,0)
# and bottomright (column,row)=(510,128) and a thickness of 3 px
cv.rectangle(img, (384,0), (510,128), (0,255,0), 3)

# Display the image (now containing a line and a rectangle)
plt.imshow(img[:,:,::-1]);
plt.xticks([]); plt.yticks([]);
plt.show()
```

## Drawing circles

The `cv.circle()` function expects the input arguments below:
- The image to draw on.
- The circle's center coordinates (column, row).
- The circle's radius in pixels (px).
- A color: (b,g,r) values for a BGR image, a scalar value for a gray scale image.
- The line thinckness in pixels (px), or -1 to fill the circle with the chosen color.

```python
# Draw a red (b,g,r=0,0,255) circle with it's center at (column,row)=(447,63),
# radius of 63 px, and fill the circle (i.e. thickness of -1 px)
cv.circle(img, (447,63), 63, (0,0,255), -1)

# Display the image (now containing a line, a rectangle, and a circle)
plt.imshow(img[:,:,::-1]);
plt.xticks([]); plt.yticks([]);
plt.show()
```

## Drawing ellipses

The `cv.ellipse()` function expects the input arguments below:
- The image to draw on.
- The ellipse's center coordinates (column, row).
- The half-length of the ellipse's axes (major/horizontal axis half-length, minor/vertical axis half-length).
- The angle (in degrees) the ellipse is rotated in the anti-clockwise direction from the major/horizontal axis.
- The starting angle (in degrees) of the ellipse's arc in the clockwise direction from the major/horizontal axis.
- The ending angle (in degrees) of the ellipse's arc in the clockwise direction from the major/horizontal axis.
- A color: (b,g,r) values for a BGR image, a scalar value for a gray scale image.
- The line thinckness in pixels (px), or -1 to fill the ellipse with the chosen color.

```python
# Draw a blue (b,g,r=255,0,0) ellipse with it's center at (column,row)=(256,256),
# major axis half-length of 100 and minor axis half-length of 50 (100,50),
# no anti-clockwise rotation from the major/horizontal axis (first 0),
# start angle of 0 degrees from the major/horizontal axis in the clockwise direction (second 0),
# end angle of 180 degrees from the major/horizontal axis in the clockwise direction (180),
# and fill the ellipse (i.e. thickness of -1 px)
cv.ellipse(img, (256,256), (100,50), 0, 0, 180, (255,0,0), -1)

# Display the image (now containing a line, a rectangle, a circle, and an ellipse)
plt.imshow(img[:,:,::-1]);
plt.xticks([]); plt.yticks([]);
plt.show()
```

## Drawing polygons

The `cv.polylines()` function expects the input arguments below:
- The image to draw on.
- An array (of type `int32`) of polygon vertices (column, row) of shape (rows, 1, 2), where *rows* is the number of vertices.
- A boolean "isClosed" flag. If True, the polygon is closed, i.e. a line is drawn from the last- to the first vertex.
- A color: (b,g,r) values for a BGR image, a scalar value for a gray scale image.
- The line thinckness in pixels (px), or -1 to fill the polygon with the chosen color.

```python
# Draw a yellow (b,g,r=0,255,255), closed (True) polygon,
# using the 4 vertices [column,row] = [10,5], [20,30], [70,20], [50,10],
# and a line thickness of 5 px (pixels)
pts = np.array([[10,5], [20,30], [70,20], [50,10]], np.int32)
pts = pts.reshape((-1,1,2))
cv.polylines(img, [pts], True, (0,255,255), 5)

# Display the image (now containing a line, a rectangle, a circle, an ellipse, and a polygon)
plt.imshow(img[:,:,::-1]);
plt.xticks([]); plt.yticks([]);
plt.show()
```

## Drawing text

The `cv.putText()` function expects the input arguments below:
- The image to draw on.
- The text (string) to draw.
- The bottom left corner (column, row) to place the text in the image.
- Font type (different font types can be found [here](https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#ga0f9314ea6e35f99bb23f29567fc16e11)).
- Font size.
- A color: (b,g,r) values for a BGR image, a scalar value for a gray scale image.
- The line thickness in pixels (px).
- line type, e.g. anti-aliased `cv.LINE_AA`

```python
# Draw the text 'OpenCV' to the image with:
# - bottom right corner at (column,row)=(10,500)
# - font "Hershey Simplex"
# - font size of 4
# - white color (255,255,255)
# - line thickness of 2 px
# - line style "AA" (anti-aliased)
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(img, 'OpenCV', (10,500), font, 4, (255,255,255), 2, cv.LINE_AA)

# Display the image (now containing a line, a rectangle, a circle, an ellipse, a polygon, and some text)
plt.imshow(img[:,:,::-1]);
plt.xticks([]); plt.yticks([]);
plt.show()
```
